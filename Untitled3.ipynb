```python
import re

# ---- Keep only relevant columns ----
# Adjust column name if different in your file
df = df[['Reports']].copy()

# ---- Drop nulls ----
df.dropna(subset=['Reports'], inplace=True)
df.reset_index(drop=True, inplace=True)

print(f"Total records after cleaning: {len(df)}")

# ---- Text Cleaning Function ----
def clean_text(text):
    text = str(text).lower()                        # Lowercase
    text = re.sub(r'http\S+|www\S+', '', text)      # Remove URLs
    text = re.sub(r'[^a-z\s]', '', text)            # Remove numbers & special chars
    text = re.sub(r'\s+', ' ', text).strip()        # Remove extra spaces
    return text

# ---- Apply Cleaning ----
df['Cleaned_Report'] = df['Reports'].apply(clean_text)

# ---- Limit size for Colab (to avoid memory issues) ----
df = df.sample(n=min(500, len(df)), random_state=42).reset_index(drop=True)

print(f"\nWorking with {len(df)} samples")
print("\n--- Sample cleaned text ---")
print(df[['Reports', 'Cleaned_Report']].head(3))
```

```python
from transformers import pipeline

# ---- Load pre-trained sentiment model ----
# DistilBERT: lightweight BERT, fast on Colab, strong accuracy
print("Loading sentiment model... (this may take 1-2 minutes)")

sentiment_pipeline = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english",
    truncation=True,
    max_length=512
)

print("Model loaded successfully!")

# ---- Run sentiment analysis in batches ----
# Batching avoids memory crashes on Colab
def get_sentiment_batch(texts, batch_size=32):
    results = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size].tolist()
        batch_results = sentiment_pipeline(batch)
        results.extend(batch_results)
        print(f"  Processed {min(i+batch_size, len(texts))}/{len(texts)} rows...", end='\r')
    print("\nDone!")
    return results

# ---- Apply to cleaned text ----
print("\nRunning sentiment analysis...")
raw_results = get_sentiment_batch(df['Cleaned_Report'])

# ---- Extract label and score ----
df['Sentiment']       = [r['label'] for r in raw_results]   # POSITIVE / NEGATIVE
df['Sentiment_Score'] = [round(r['score'], 4) for r in raw_results]  # Confidence

# ---- Map to friendlier labels ----
df['Sentiment'] = df['Sentiment'].map({'POSITIVE': 'Positive', 'NEGATIVE': 'Negative'})

# ---- Preview results ----
print("\n--- Sample Sentiment Results ---")
print(df[['Cleaned_Report', 'Sentiment', 'Sentiment_Score']].head(6).to_string())

print("\n--- Sentiment Distribution ---")
print(df['Sentiment'].value_counts())
```

```python
from transformers import pipeline as hf_pipeline

# ---- Define complaint categories ----
CATEGORIES = [
    "Academic",
    "Library & Resources",
    "Technology & Software",
    "Faculty & Teaching",
    "Administration & Fees",
    "Campus Facilities",
    "International Students",
    "Mental Health & Support"
]

# ---- Load zero-shot classification model ----
print("Loading classification model... (may take 1-2 minutes)")

classifier = hf_pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=-1  # CPU mode for Colab (use 0 for GPU if available)
)

print("Model loaded!\n")

# ---- Run classification in batches ----
def classify_batch(texts, categories, batch_size=16):
    labels = []
    scores = []
    total = len(texts)
    for i in range(0, total, batch_size):
        batch = texts[i:i+batch_size].tolist()
        for text in batch:
            result = classifier(text, candidate_labels=categories, truncation=True)
            labels.append(result['labels'][0])        # Top predicted category
            scores.append(round(result['scores'][0], 4))  # Confidence score
        print(f"  Classified {min(i+batch_size, total)}/{total} rows...", end='\r')
    print("\nDone!")
    return labels, scores

# ---- Apply to cleaned complaints ----
print("Running category classification (this takes ~3-5 mins on CPU)...")
cat_labels, cat_scores = classify_batch(df['Cleaned_Report'], CATEGORIES)

# ---- Store results ----
df['Category']       = cat_labels
df['Category_Score'] = cat_scores

# ---- Preview results ----
print("\n--- Sample Classification Results ---")
print(df[['Cleaned_Report', 'Category', 'Category_Score']].head(6).to_string())

print("\n--- Category Distribution ---")
print(df['Category'].value_counts())
```

```python
print("=" * 60)
print("   SMART STUDENT FEEDBACK INSIGHT SYSTEM")
print("   Structured Analysis Report")
print("=" * 60)

total = len(df)
neg = (df['Sentiment'] == 'Negative').sum()
pos = (df['Sentiment'] == 'Positive').sum()
top_cat = df['Category'].value_counts().idxmax()
avg_conf = df['Category_Score'].mean()

print(f"\nTotal Complaints Analyzed : {total}")
print(f"Negative Sentiment        : {neg} ({neg/total*100:.1f}%)")
print(f"Positive Sentiment        : {pos} ({pos/total*100:.1f}%)")
print(f"Most Common Category      : {top_cat}")
print(f"Avg Classification Conf.  : {avg_conf:.2%}")

print("\n--- Full Structured DataFrame (first 10 rows) ---")
output_df = df[['Reports', 'Sentiment', 'Sentiment_Score', 'Category', 'Category_Score']].copy()
output_df.columns = ['Original Report', 'Sentiment', 'Sent. Confidence',
                     'Category', 'Cat. Confidence']
print(output_df.head(10).to_string())

# Save to CSV
output_df.to_csv('feedback_insights.csv', index=False)
print("\nResults saved to: feedback_insights.csv")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import numpy as np

# ---- Group complaints by category ----
category_texts = df.groupby('Category')['Cleaned_Report'].apply(lambda x: ' '.join(x))

# ---- Fit TF-IDF on all categories ----
vectorizer = TfidfVectorizer(
    max_features=500,
    stop_words='english',
    ngram_range=(1, 2)      # single words AND two-word phrases
)
tfidf_matrix = vectorizer.fit_transform(category_texts)
feature_names = vectorizer.get_feature_names_out()

# ---- Extract top N keywords per category ----
TOP_N = 6

print("=" * 60)
print("  TOP KEYWORDS PER COMPLAINT CATEGORY")
print("=" * 60)

category_keywords = {}
for i, category in enumerate(category_texts.index):
    scores = tfidf_matrix[i].toarray()[0]
    top_indices = scores.argsort()[::-1][:TOP_N]
    top_keywords = [(feature_names[j], round(scores[j], 4)) for j in top_indices if scores[j] > 0]
    category_keywords[category] = top_keywords

    print(f"\n{category}")
    for kw, score in top_keywords:
        bar = '█' * int(score * 40)
        print(f"   {kw:<30} {bar} ({score})")

print("\n")
```
